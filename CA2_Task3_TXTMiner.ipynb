## Creating wordCloud for each type of wine (41 wines)
As you can see, we can check how many wine varieties are present in the dataset with the aid of the unique function. Simply, we are storing the result of the unique() function in wineTypes variable and checking the length, although the len() function has returned 41 as a value. Therefore, we need to display 41 wordclouds with the most used words in the description type of wine.   

```python
wineTypes= np.unique(df['variety'])
wineTypes
```

```python
len(wineTypes)
```

Here we have created a function such as generateWordCloud which iterates 41 times because we have a total 41 wine types and find the record based on wine type. Besides, description is being converted into a string and passed into wordcloud to generate wordcloud for each wine type. Here, we see that the wordcloud function requires the max words parameter to produce the greatest number of words from the string. The imshow, 2-dimensional Numpy array is used to make a picture using the matplotlib function imshow(). Each square in the image will represent each element in the array. The value of the relevant array element and the color map used by imshow() and together define the color of each square. Bilinear interpolation has been used to display a smoother image.

```python
def generateWordCloud():
  for types in wineTypes:
    variety = df[df['variety'].str.contains(types)]
    description = str(variety['description'])
    print('WordCloud to the wine:',types)
    wordcloud = WordCloud(max_words=5).generate(description)
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.show()
    print()

generateWordCloud()
```

## Creating wordCloud for all records (3057)

### Creating a function to get most 5 used Words from each description
Here, we've developed a program to select the most often used words based on their frequency. Wordcould's process text function has been called to display the word count and relative counts from the text, respectively. In addition, iloc gives us the ability to choose a specific row or column of the dataset. In the next line, we are sorting the dictionary elements that will be returned with the aid of the key as lambda, which enables the creation of short, anonymous, and single-line functions that return the value of an expression. A blank list has been made to hold the first five words, and it is then periodically expanded to include the next five words. When checking the five common words from the first records, the function returns the exact same words that we displayed in above wordCloud for single record

```python
commonWords = []

# creating a function to
def wordGenerator():
  for index in range(len(df['description'])):
    text_dictionary = wordcloud.process_text(df['description'].iloc[index])
    word_freq={k: v for k, v in sorted(text_dictionary.items(),reverse=True, key=lambda item: item[1])}
    commonWords.extend(list(word_freq.keys())[:5])


wordGenerator()

```

We have total 3057 rows in dataset and we need to pick 5 most used words from each records.

i.e = 3057 X 5 = 15,285 words

As we can see, the above function produced exactly 15285 words. We used the len() method to calculate the total number of words in the list.

```python
len(commonWords)
```

### Converting list of words into String
To turn each word into a string in this cell, we must iterate each word. A blank string was made and then each word was added one after the other. Therefore, it is clear that the print statement shows the longest string, which has 15285 words.

```python

commonWordString = ''

for word in commonWords:
  commonWordString += str(word + " ")
print("Word of String = ", commonWordString)
```

### WordCloud for All records with most common used words from each description
Height, width, background color, and Stopwords are additional parameters that have been added to the wordcloud() function. Using Height and Width, we may modify the canvas's dimensions. Here, we've given the dimensions 1600 width by 800 hight.The word cloud image's background color is white. if the canvas's dimensions are 800 by 600. We might lose canvas quality, however using the figsize function is an easy workaround. Use the tight layout() function with pad set to 0 to reduce the size of the border. The bilinear was used to make images more fluid.

```python
stopwords = set(STOPWORDS)
wordcloud = WordCloud(stopwords=stopwords,width=1600, height=800,background_color="white",max_words=15285).generate(commonWordString)
print(wordcloud)
plt.figure(figsize=(10,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```

## Masking to display words in particular shape
Our Dataset is related to wine so, here we are tryng to display glass shape by using mask concept.Two libraries, such as Image and request, will be need to mask the image with the wordcloud. To open an image from a URL, use the image library. To fetch an image over HTTP, use the request package.By default, a rectangle is used in word clouds. Given that we constructed a variable in the form of a picture, we should give it to the mask argument.

```python
 from PIL import Image
 import requests
```

```python
from PIL import Image
stopwords = set(STOPWORDS)
pic = np.array(Image.open(requests.get('http://www.clker.com/cliparts/4/3/8/e/12085302761699332124coredump_Simple_wine_glass.svg.med.png',stream=True).raw))
wordcloud = WordCloud(stopwords=stopwords,width=1600, height=800, background_color="white", mask=pic).generate(commonWordString)
plt.figure(figsize=(7,7))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
